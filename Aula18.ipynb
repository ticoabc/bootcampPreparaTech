## **Estatística Descritiva**

As estatísticas descritivas envolvem os métodos usados para resumir, simplificar e descrever os principais recursos de um conjunto de dados.

Este ramo da estatística visa fornecer uma visão geral concisa e significativa dos dados por meio de medidas de tendência central (média, mediana, moda), dispersão (intervalos, variância, desvio-padrão) e representação gráfica, tais como histogramas, gráfico de barras, gráfico de setores, gráficos de dispersão e etc.


As estatísticas descritivas nos ajudam a compreender as caracteristicas fundamentais de um conjunto de dados e torná-lo mais gerenciável para análise.

###Tradução/Renomear###

```python
# Renomeando colunas para tradução
df.rename(columns={'DATA_BASE': 'data_base',
                   'FAIXA_DESENROLA': 'faixa',
                   'UNIDADE_FEDERACAO':'uf',
                   'COD_CONGLOMERADO_FINANCEIRO': 'cod_cng_fin',
                   'NOME_CONGLOMERADO_FINANCEIRO': 'nome_inst_fin',
                   'NUMERO_OPERACOES': 'qtd_operacoes',
                   'VOLUME_OPERACOES': 'vol_operacoes',
                   }, inplace=True)
```

```python
# Checando alteração
df.info()
```

```python
# Modo 1: verificando valores nulos
df.isna().sum()
```

```python
# Modo 2: verificando valores nulos
df.isnull().sum()
```

```python
# Modo 3: Calcular a porcentagem de valores nulos em cada coluna
porcentagem_nulos = (df.isnull().sum() / len(df)) * 100

# Exibir a porcentagem de valores nulos em cada coluna
print(round(porcentagem_nulos,1))
```

```python
# Removendo linhas com valores NaN
df = df.dropna()
```

```python
# Checagem se de fato os valores nulos foram eliminados
df.isna().sum()
```

```python
# Comando para verificar o balanceamento do dataframe
df.info()
```
